```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1. Read the dataset to python environment
iris_data = pd.read_excel('/content/iris.xls')
iris_data.head()
```

```python
#Information
iris_data.info()
```

```python
#2. Display the columns in the dataset
iris_data.columns
```

```python
#3.Calculate the mean of each column of the dataset.(Method 1)
#Describe
iris_data.describe()
```

```python
#3.Calculate the mean of each column of the dataset.(Method 2)
mean = iris_data.mean(numeric_only=True)
mean
```

```python
# 4.Check for the null values present in the dataset
iris_data.isna().sum()
```

```python
# 5. Perform meaningful visualizations using the dataset.
```

```python
# Finding unique values and value counts in classification
iris_data["Classification"].value_counts()
```

```python
#number of counts in each classification
sns.countplot(x='Classification',hue='Classification',data=iris_data,width = 0.5)
plt.show()
```

```python
#Bar Chart - Classification v/s PW
plt.bar(iris_data['Classification'],iris_data['PW'],width=0.5)
plt.title('Classification v/s PW')
plt.xlabel('Classification')
plt.ylabel('PW')
plt.show()
```

```python
#Scatter plots
```

```python
#PL v/s PW
sns.scatterplot(x=iris_data['PL'],y=iris_data['PW'],hue=iris_data['Classification'])

```

```python
#SL v/s SW
sns.scatterplot(x=iris_data['SL'],y=iris_data['SW'],hue=iris_data['Classification'])
```

```python
#Boxplots
fig, axs = plt.subplots(2, 2, figsize=(8, 8))
sns.boxplot(x='Classification', y='SW', data=iris_data, ax=axs[0, 0])
axs[0, 0].set_title('SW')

sns.boxplot(x='Classification', y='SL', data=iris_data, ax=axs[0, 1])
axs[0, 1].set_title('SL')

sns.boxplot(x='Classification', y='PW', data=iris_data, ax=axs[1, 0])
axs[1, 0].set_title('PW')

sns.boxplot(x='Classification', y='PL', data=iris_data, ax=axs[1, 1])
axs[1, 1].set_title('PL')
plt.tight_layout()
```
