```python
#outliers in 'Malic_Acid', 'Ash 'Ash_Alcanity', 'Magnesium''Proanthocyanins', 'Color_Intensity', 'Hue'
```

```python
features = ['Malic_Acid', 'Ash' ,'Ash_Alcanity', 'Magnesium','Proanthocyanins', 'Color_Intensity', 'Hue']

# Function to detect and cap outliers for a given feature
def handle_outliers(df, feature):
    q1 = df[feature].quantile(0.25)
    q3 = df[feature].quantile(0.75)
    iqr = q3 - q1
    low_lim = q1 - 1.5 * iqr
    upp_lim = q3 + 1.5 * iqr

    outliers = df[feature][(df[feature] < low_lim) | (df[feature] > upp_lim)]
    df[feature] = df[feature].clip(lower=low_lim, upper=upp_lim)

    return outliers, low_lim, upp_lim

# Iterate over each feature to handle outliers
for feature in features:
    outliers, low_lim, upp_lim = handle_outliers(data, feature)
    print(f'Feature: {feature}')
    print(f'  Q1: {data[feature].quantile(0.25)}')
    print(f'  Q3: {data[feature].quantile(0.75)}')
    print(f'  IQR: {data[feature].quantile(0.75) - data[feature].quantile(0.25)}')
    print(f'  Lower limit: {low_lim}')
    print(f'  Upper limit: {upp_lim}')
    print(f'  Outliers: {outliers.tolist()}')

    # Plotting the boxplot for each feature
    plt.figure()
    plt.boxplot(data[feature])
    plt.title(f'Boxplot for {feature}')
    plt.show()
```

**Kmeans**

```python
from sklearn.cluster import KMeans
```

```python
#find wcss and use elbow method to find optimum number of cluster
wcss = []

for i in range(1,11):
  kmeans = KMeans(n_clusters = i , init = 'k-means++', max_iter=300 ,
                  n_init =10 , random_state =42)
  kmeans.fit(data)
  wcss.append(kmeans.inertia_)

wcss
```

```python
#elbow method
plt.figure(figsize=(6,6))
plt.plot(range(1,11),wcss,linewidth=2,color="red",marker="o")
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel("Wcss")
plt.xticks(np.arange(1,11,1))
plt.show()
```

```python
#take number of clusters = 3
kmeans = KMeans(n_clusters = 3 , init = 'k-means++', max_iter=300 ,
                  n_init =10 , random_state =42)
kmeans.fit(data)

cluster_kmeans = kmeans.predict(data)

cluster_kmeans
```

```python
kmeans.cluster_centers_
```

```python
#Add a new columns showng clusters found by kmeans
data_cluster = data.copy()

data_cluster['Cluster_kmeans'] =cluster_kmeans
data_cluster.head()
```

```python
data_cluster['Cluster_kmeans'].value_counts()
```

**Agglomerative** **Clustering**

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
#creating dendrogram to find number of clusters
mergings=linkage(data,method='ward')
dendrogram(mergings)
plt.title('Dendrogram')
plt.show()
```

```python
#take number of clusters =3
from sklearn.cluster import AgglomerativeClustering
agg_hc=AgglomerativeClustering(n_clusters=3,metric='euclidean',linkage='ward')
y_agg_hc=agg_hc.fit_predict(data)
y_agg_hc
```

```python
#creating new column to show clusters formed by agglomerative method
data_cluster['Cluster_agglo']=y_agg_hc
data_cluster.head()
```

```python
data_cluster['Cluster_agglo'].value_counts()
```
